```python
model_data=df[(df.AgencyName=='Citywide Administrative Services')&(df.CategoryDescription=='Goods')]
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
from math import log10,exp
import matplotlib.pyplot as plt
from statsmodels.api import OLS
from scipy.stats import chisquare
import datetime
from datetime import datetime
import warnings
warnings.filterwarnings("ignore")
pd.options.display.max_columns = None
```

```python
def next_move(coordinate):
    x=coordinate[0]
    y=coordinate[1]
    my_dic={str(x+2)+','+str(y+1):1,str(x+1)+','+str(y+2):1}
    if y>=1:
        my_dic.update({str(x+2)+','+str(y-1):1})
        if y>=2:
            my_dic.update({str(x+1)+','+str(y-2):1})
            if x>=1:
                my_dic.update({str(x-1)+','+str(y-2):1})
                
        if x>=2:
            my_dic.update({str(x-2)+','+str(y-1):1})
    if x>=1:       
        my_dic.update({str(x-1)+','+str(y+2):1})
        if x>=2:
            my_dic.update({str(x-2)+','+str(y+1):1})
    next_location= random.choice(list(my_dic.keys()))     
    next_location=next_location.split(',')
    next_coor=[int(next_location[0]),int(next_location[1])]        
    return next_coor
```

```python
next_move([0,1])
```

```python
distance=0
std=0
for i in range(500000):
    my_coor=[0,0]
    for i in range(10):
        my_coor=next_move(my_coor)
    distance=distance+np.sqrt(my_coor[0]**2+my_coor[1]**2)
    std=std+my_coor[0]**2+my_coor[1]**2
distance/500000,np.sqrt(std/500000-(distance/500000)**2)
```

```python
distance=0
std=0
count=0
for i in range(500000):
    my_coor=[0,0]
    for i in range(10):
        my_coor=next_move(my_coor)
    sample_dis=np.sqrt(my_coor[0]**2+my_coor[1]**2)
    if  sample_dis>10:   
        distance=distance+sample_dis
        std=std+sample_dis**2
        count=count+1
distance/count,np.sqrt(std/count-(distance/count)**2)
```
